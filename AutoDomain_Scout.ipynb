```python
#1. Install necessary libraries
!pip install python-whois dnspython termcolor

import whois
import dns.resolver
from termcolor import colored

print(colored("============================================","green"))
print(colored("   [+] Automated OSINT Tool - phase 1 loaded ","green", attrs=['bold']))
print(colored("===========================================","green"))

#Request input from the user
target_domain = input("\nEnter Target Domain (e.g., google.com): ")

#--------------------------------------------------------------
# [1] WHoIS Lookup (Domain Ownership dateils)
#--------------------------------------------------------------
print(colored(f"\n[*]fetching WHOIS date for: {target_domain}...","blue"))
try:
  domain_info = whois.whois(target_domain)

  print(colored("\n[+] WHOIS Results:", "green", attrs=['bold']))
  print(f"    - Registrar: {domain_info.registrar}")
  print(f"    - Creation Date: {domain_info.creation_date}")
  print(f"    -Expiration Date: {domain_info.expiration_date}")
  print(f"    -Country: {domain_info.country}")
except Exception as e :
    print(colored(f"[-]Could not fetch WHOIS date: {e}", "red"))

    #---------------------------------------------------------
    #[2] DNS Enumertion (Network Records)
    #---------------------------------------------------------
    print(colored(f"\n[*] Querying DNS Records: for: {target_domain}...", "blue"))

    record_types = ['A', 'MX', 'TXT']

    for record in record_types:
      print(colored(f"\n[+] {record} Records:", "cyan", attrs=['bold']))
try:
        answers = dns.resolver.resolves(target_domain, record)
        for rdate in answers:
          print(f"      -> {rdate.to_text()}")
except dns.resolver.NoAnswer:
            print(colored(f"       [-] No {record} records found.", "yellow"))
except Exception as e:
          print(colored(f"     [-] Error fetching {record} records.", "red"))

          print(colored("\n======================================", "green"))
          print(colored("    [+] Scan finished Successfully!    ", "green", attrs=['bold']))
          print(colored("=======================================", "green"))


```
