```python
import re
def check_password_strength(password):
  length_error = len(password) < 8
  digit_error = re.search(r"[0-9]", password) is None
  uppercase_error = re.search(r"[A-Z]", password) is None
  symbol_error = re.search(r"[ !@#$%^&*()_+={}\[\]:;<>,.?~\\/-]", password) is None
  errors = [length_error, digit_error, uppercase_error, symbol_error]
  num_errors = sum(errors)
  print("/n--- password Security Report ---")
  if num_errors == 0:
    print("Password Strength: Strong")
  elif num_errors == 1:
    print("password Strength: Medium")
  else:
    print("Password Strength: Weak")
    if length_error:
      print("Alert: password is too short (min 8 characters).")
      if digit_error:
        print("Alert: Missing at least one digit (0-9).")
        if uppercase_error:
          print("Alert: Missing at least one uppercase letter (A-Z).")
          if symbol_error:
            Print("Alert: Missing at least one special character.")
            user_input = input("Enter password to test: ")
            check_password_strength(user_input)
```

```python
user_input = input("Enter password to test: ")
check_password_strength(user_input)
```
