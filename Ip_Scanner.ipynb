```python
import socket
for host in ["google.com", "facebook.com"]:
    try:
        ip = socket.gethostbyname(host)
        print(f"[ONLINe] {host} ---> {ip}")
    except:
      print(f"[ERROR] {host} ---> Cannot find IP")
```
